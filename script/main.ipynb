```python
import deduplicate
```

```python
def de_duplicate_icd9(input_code : list[str]) -> list[str]:
    code = deduplicate.remove_invalid_code(input_code)
    # If there is more detailed code -> remove less digit code
    dedup_code = deduplicate.remove_less_detailed_code(code)
    # If code has include code -> remove include code
    dedup_code = deduplicate.remove_include_code(dedup_code)
    # Collapse two code into one code
    dedup_code = deduplicate.collapse_combine_code(dedup_code)
    
    deduplicate.alert_reconsider_code(dedup_code)
    return dedup_code
```

```python
input_code = ["8331", "833",
              "8460", "8051",
              "1733", "1735",
              "0056", "0057", # This code is exceptional since it require each other codeAlso
              "0044", "0045"
              ]

de_duplicate_icd9(input_code=input_code)
```
